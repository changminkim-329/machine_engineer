- [참조](/notebooks/백엔드/프록시%20서버.ipynb#포워드-프록시(Forward-Proxy))
- 클라이언트가 Proxy 서버에 요청을 하면, 프록시 서버가 배후 서버(REVERSE server)서버로부터 데이터를 가져오는 역할을 함.
- 프록시 서버: NGINX, 리버스 서버: 어플리케이션 서버,django, flask

- 웹 응용프로그램 서버에 리버스 프록시(Nginx)를 두는 이유는 요청(request)에 대한 버퍼링이 있기 때문
    - 클라이언트가 직접 어플리케이션 서버에 직접 요청하는 경우, 프로세스 1개가 응답 대기 상태가 되어야 한다. 따라서 프록시 서버를 둠으로써 요청을 배분하는 역할을 한다.
    - **nginx.conf** 파일에 location지시어를 사용하여 요청을 배분.

- Nginx는 비동기 처리 방식(Event-Drive) 방식을 채택
    - 동기(Synchronous) : A가 B에게 데이터를 요청했을 때, 이 요청에 따른 응답을 주어야만 A가 다시 작업 처리가 가능 (하나의 요청, 하나의 작업에 충실)
    - 비동기(Asynchronous) : A의 요청을 B가 즉시 주지 않아도, A의 유휴시간으로 또 다른 작업 처리가 가능한 방식

### Nginx 디렉토리 구조 확인
- nginx 디렉토리 경로: **/etc/nginx**

#### 디렉토리 내부 파일 과 디렉토리
- conf.d(디렉토리) : nginx.conf에서 불러들일 수 있는 파일을 저장.
- proxy.conf: 프록시 관련 환경 설정.
- fastcgi.conf(파일) : FastCGI 환경설정 파일
- nginx.conf(파일) : 접속자 수, 동작 프로세스 수 등 퍼포먼스에 관한 설정들.
- sites-avaliable(디렉토리) : 비활성된 사이트들의 설정 파일들이 위치.
- sties-enabled(디렉토리) : 활성화된 사이트들의 설정 파일들이 위치.

```python
user www-data; 
# user: NGINX 프로세스가 실행되는 권한 지정(디폴트 값: www-data)
# 만약 user의 값이 root로 되어 있다면 워커 프로세스를 root의 권한으로 동작.
# 워커 프로세스를 악의적 사용자가 제어 된다면 루트 사용자의 권한으로 원격제어하게 되기 때문에 위험
# 이 설정의 값으로는 www-data, www, nginnx와 값이 계정이 하는 일에 대한 
# 대표성이 있는 이름을 사용하면 되고, 이 계정은 일반 유저의 권한으로 쉘에 접속 할 수 없어야 안전.
## 예를들어 지시자의 값이 user www-data라면 아래와 같이 사용자를 생성
## useradd - shell /usr/sbin/nologin www-data 
## 이렇게 하면 www-data의 구한으로 

## nginx는 마스터(master)와 워커(worker)프로세스로 나뉨.
### 워커 프로세스가 실질적인 웹서버 역할을 수행.
## user 지시어는 워커 프로세스의 권한을 지정.


worker_processes auto;
# work_processes 지시어: 워커 프로세스를 몇개 생성할 것인지를 지정하는 지시어.
## 위에서 언급한 워커 프로세스 이다.
## 실질적인 웹서버 역할을 한다.
## 이 값이 1이라면 모든 요청을 하나의 프로세스로 실행하겠다는 의미로
## 여러개의 CPU 코어가 여러개 있는 시스템에서 이 값이 1이면 하나의 코어만으로 요청을 처리
## auto도 무방하지만, 명시적으로 서버에 장착되어 있는 코어 수 만큼 할당하는 것이 보통이며,
## 더 높게도 설정 가능.
# work_processes : NGINX 프로세스 실행 가능 수



pid /run/nginx.pid;
# NGINX 마스터 프로세스 ID 정보가 저장.

include /etc/nginx/modules-enabled/*.conf;
# 포함 시킬 외부 파일을 정의한다. 


events {
        worker_connections 768;
        # multi_accept on;
}
## 주로 네트워크 동작에 관련된 설정을 하는 영역
## worker_connections: Worker Process가 동시에 처리할 수 있는 접속자의 수 나타냄.
## worker_processes 1024 : 1024 접속사 동시에 처리 가능. 

# worker_process 값이 4 이고, worker_connnections의 값이 1024라면 
# 4 x 1024 = 4096개의 커넥션을 맺을 수 있음.


# nginx에는 http와 관련된 설정을 하는 영역
http {

        ##
        # Basic Settings
        ##

        sendfile on;
        tcp_nopush on;
        tcp_nodelay on;
        keepalive_timeout 65;
        types_hash_max_size 2048;
        # server_tokens off;

        # server_names_hash_bucket_size 64;
        # server_name_in_redirect off;

        include /etc/nginx/mime.types;
        default_type application/octet-stream;

        ##
        # SSL Settings
        ##

        ssl_protocols TLSv1 TLSv1.1 TLSv1.2; # Dropping SSLv3, ref: POODLE
        ssl_prefer_server_ciphers on;

        ##
        # Logging Settings
        ##

        access_log /var/log/nginx/access.log;
        error_log /var/log/nginx/error.log;
        # error_log: 로그 레벨을 설정하는 지시어
        ## 로그레벨은 [debug|info|notice|warn|error|crit]

        ##
        # Gzip Settings
        ##
        
    
        gzip on;

        # gzip_vary on;
        # gzip_proxied any;
        # gzip_comp_level 6;
        # gzip_buffers 16 8k;
        # gzip_http_version 1.1;
        # gzip_types text/plain text/css application/json application/javascript text/xml application/xml application/xml+rss text/javascript;

        ##
        # Virtual Host Configs
        ##

        include /etc/nginx/conf.d/*.conf;
        include /etc/nginx/sites-enabled/*;
}


#mail {
#       # See sample authentication script at:
#       # http://wiki.nginx.org/ImapAuthenticateWithApachePhpScript
#
#       # auth_http localhost/auth.php;
#       # pop3_capabilities "TOP" "USER";
#       # imap_capabilities "IMAP4rev1" "UIDPLUS";
#
#       server {
#               listen     localhost:110;
#               protocol   pop3;
#               proxy      on;
#       }
#
#       server {
#               listen     localhost:143;
#               protocol   imap;
#               proxy      on;
#       }
#}
                                              
```
